```python
import pandas as pd
import requests
from io import StringIO
```

```python
url = 'https://en.wikipedia.org/wiki/World_population'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

data = StringIO(response.text)

tables = pd.read_html(data, match='Ten most populous countries')
```

```python
# if want to get csv file from website use pd.read_csv('url')
```

```python
df = tables[0]
df.to_csv('world_pop.csv', index=False)
```

```python
tables1 = pd.read_html(data, match='10 most densely populated countries')
df1 = tables1[0]
df1.rename(columns={'Rank': 'Country Rank'}, inplace=True)
df1.to_csv('world_dense_pop.csv', index=False)
```
