```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_brightest_stars'
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
print(response)

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
all_tables = soup.find_all('table', class_='wikitable')
# print(all_tables)

target_table = all_tables[2]
print(target_table)

# # Find a table that has a specific width attribute
# table = soup.find('table', attrs={'style': 'width:100%'})

# # OR find a table by its ID (IDs are always unique)
# table = soup.find('table', id='customer_data')
```

```python
heads = target_table.find_all('th')
titles = [i.get_text(strip=True) for i in heads]

all_rows = []

rows = target_table.find_all('tr')

for i in rows[1:]:
    data = i.find_all(['td', 'th'])

    row = [tr.get_text(strip=True) for tr in data]

    if len(row) == len(titles):
        all_rows.append(row)

df = pd.DataFrame(all_rows, columns=titles)
print(df.head())

df.to_csv('Star Data.csv', index=False)
```
